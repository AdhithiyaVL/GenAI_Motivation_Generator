```python
from transformers import pipeline, set_seed
```

```python
generator = pipeline('text-generation', model='gpt2')  # or 'gpt2-medium' for better results
sentiment_analyzer = pipeline("sentiment-analysis")
set_seed(42)  
```

```python
def generate_motivation(prompt="Give me a motivational quote"):
    print("\n Generating Motivational Quote...")
    results = generator(prompt, max_length=50, num_return_sequences=1)
    generated_text = results[0]['generated_text']
    
    print("\n Motivational Quote:")
    print(generated_text.strip())

    return generated_text.strip()

def analyze_sentiment(text):
    print("\n Analyzing Sentiment...")
    result = sentiment_analyzer(text)[0]
    print(f" Sentiment: {result['label']}")
    print(f" Confidence: {round(result['score'] * 100, 2)}%")

# Main
if __name__ == "__main__":
    quote = generate_motivation()
    analyze_sentiment(quote)
```
